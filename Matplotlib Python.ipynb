```python
from matplotlib import pyplot as plt         ## importing matplotlib library
```

```python
x_values = [10,11,12,13]
y_values = [20,21,22,23]
```

### Line Plot

```python
plt.plot(x_values, y_values)
plt.show()
```

### Scatter Plot

```python
plt.scatter(x_values, y_values)
plt.show()
```

### Combine Plotting    

```python
other_x_values = [10,11,12,13]
other_y_values = [31,32,33,34]
```

```python
plt.plot(other_x_values, other_y_values)
plt.title("Sample Plot")
```

```python
plt.plot(other_x_values, other_y_values)
plt.title("Sample Plot")
plt.xlabel("x values")
plt.ylabel("y values")
```

```python
plt.plot(other_x_values, other_y_values, color="red")
plt.title("Sample Plot")
plt.xlabel("x values")
plt.ylabel("y values")
```
